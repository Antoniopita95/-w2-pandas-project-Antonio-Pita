```python
patron_Activity = r"([a-zA-Z]*ing)"
#Utilizamos regex para filtrar las actividades que nos puedan servir como muestra
```

```python
def coincidencia (patron, string):
    import re
    dni = re.search(patron, string)
    if dni is None:
        return 0
    return dni.group().lower()
```

```python
df["Activity"] = df["Activity"].astype(str)
```

```python
df['ACTIVITY'] = df.Activity.apply(lambda x: lt.coincidencia(patron_Activity, x))
df.head()
```

```python
df.Age.value_counts()
```

```python
df["Age"] = df["Age"].astype(str)
```

```python
patron_Age = r"(\d{2})"
#Volvemos a utilizar regex para limpiar la columna de edades
```

```python
df['AGE'] = df.Age.apply(lambda x: lt.coincidencia(patron_Age, x))
df.sample(10)
```
